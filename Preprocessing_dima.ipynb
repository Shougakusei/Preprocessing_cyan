```python
df = df[['Количество комнат', 'Метро', 'Цена', 'Окна']]
df
```

```python
df['Количество комнат'].isnull().values.sum()
```

```python
df = df.assign(rooms = df['Количество комнат'].str[0].astype(float))
df['rooms'].value_counts()
```

```python
#создаём категориальные значения кол-ва комнат
df = df.assign(rooms = df['Количество комнат'].str[0].astype(float))
df['one_room'] = [1  if x == 1.0 else 0 for x in df['rooms']] 
df['two_room'] = [1  if x == 2.0 else 0 for x in df['rooms']]
df['three_room'] = [1  if x == 3.0 else 0 for x in df['rooms']]
df['four_room'] = [1  if x == 4.0 else 0 for x in df['rooms']]
df['five_room'] = [1  if x == 5.0 else 0 for x in df['rooms']]
df['six_room'] = [1  if x == 6.0 else 0 for x in df['rooms']]
df['rooms'] = df['rooms'].fillna(0)
# df.drop('NaN_room',axis =1)
df['NaN_room'] = [1 if x == 0 else 0 for x in df['rooms']]
df
# df.drop(['Количество комнат', 'rooms'],axis = 1)
# np.NaN
```

```python
cost = df['Цена'].str.split(',', expand=True)
# print(cost)
df['rent_cost'] = cost[0].str.split(' ')
df['rent_cost'] = [float(x[0]) for x in df['rent_cost']]
# df['rent_cost'].dtype
df
```

```python
df['subway'] = df['Метро'].str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
# time_to_subway = [s for s in df['subway'] if s.isdigit()]
# time_to_subway
df['subway'] = df['subway'].fillna(0)
df['NaN_subway'] = [1 if x == 0 else 0 for x in df['subway']]
```

```python
df['subway'] = df['Метро'].str.split('(', expand=True)[1].str.split(" ", expand=True)[0]
# df['subway']= sub_time.astype(float)
# df['subway'] = df['subway'].fillna(0)
df['nan_sub'] = []
df
```

```python
df['Окна'].value_counts()
```

```python
df['dvor'] = [1 if x == 'Во двор' else 0 for x in df['Окна']]
df['ulitsa'] = [1 if x == 'На улицу' else 0 for x in df['Окна']]
df['ulitsa_dvor'] = [1 if x == 'На улицу и двор' else 0 for x in df['Окна']]
df['Окна'] = df['Окна'].fillna(0)
df['NaN_okna'] = [1 if x == 0 else 0 for x in df['Окна']]
```
