```python
def district(text5):
    res = ''
    if text5 in VAO:
        res = 'ВАО'
    elif text5 in ZAO:
        res = 'ЗАО'
    elif text5 in NAO:
        res = 'НАО'
    elif text5 in SAO:
        res = 'САО'
    elif text5 in SVAO:
        res = 'СВАО'
    elif text5 in SZAO:
        res = 'СЗАО'
    elif text5 in CAO:
        res = 'ЦАО'
    elif text5 in UVAO:
        res = 'ЮВАО'
    elif text5 in UAO:
        res = 'ЮАО'
    elif text5 in UZAO:
        res = 'ЮЗАО'
    elif text5 in PODMOSKOVIE:
        res = 'Подмосковье'
    return res
```
